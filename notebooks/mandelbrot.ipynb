```python
import sys
import pathlib
from matplotlib import pyplot
%matplotlib inline
```

```python
rootdir = pathlib.Path().resolve().parent
moduledir = rootdir / 'scripts'
if str(moduledir) not in sys.path:
    sys.path.append(str(moduledir))
```

```python
from helper import mandelbrot, mandelbrot_set
```

```python
xmin, xmax = -2.0, 0.5
ymin, ymax = -1.25, 1.25
width, height = 500, 400
maxiter = 40
x, y, n = mandelbrot_set(xmin, xmax, ymin, ymax,
                         width, height, maxiter)
```

```python
pyplot.rc('font', family='serif', size=16)
fig, ax = pyplot.subplots(figsize=(8.0, 8.0))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.imshow(n, extent=[xmin, xmax, ymin, ymax])
ax.axis('scaled', adjustable='box')
fig.tight_layout()

figdir = rootdir / 'figures'
figdir.mkdir(parents=True, exist_ok=True)
filepath = figdir / 'figure1.png'
fig.savefig(str(filepath), dpi=300)
```
